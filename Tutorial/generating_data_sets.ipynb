```python
import pandas as pd
import numpy as np

np.random.seed(32)

X = np.random.randint(low=[200,1900,1],high=[2401,2025,13], size=(200, 3))

y_high = 8
x_high = np.array([2401, 2025, 13])
x_ratio = np.array([0.6, 0.1, 0.3])

weights = (y_high * x_ratio) / x_high 

Y = X @ weights 

Y = Y + np.random.uniform(low=-1,high=1,size=(200,)) 
Y = Y.round( decimals=3)
print(Y.max())
print(Y.min())

import pandas as pd
XY = np.column_stack([X, Y])
house = pd.DataFrame(XY, columns=["sqft","year","rooms","price"])
house.to_csv('house.csv', index=False, header=True)
house

```
